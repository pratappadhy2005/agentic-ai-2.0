### Data ingestion or data loader
#https://python.langchain.com/docs/integrations/document_loaders/


```python
### Text loader
from langchain.document_loaders import TextLoader
```

```python
loader = TextLoader('sample.txt')
test_documents = loader.load()
print(test_documents[0].page_content)
```

```python
## PDF Loader
from langchain.document_loaders import PyPDFLoader

loader = PyPDFLoader('test.pdf')
pdf_documents = loader.load()
print(pdf_documents[0].page_content)
```

```python
## Web loader
from langchain.document_loaders import WebBaseLoader

loader = WebBaseLoader("https://www.gutenberg.org/cache/epub/1112/pg1112.txt")
web_documents = loader.load()
print(web_documents[0].page_content)
```

```python
from langchain.document_loaders import WikipediaLoader
loader = WikipediaLoader(query="2024 India World Cup")
wiki_documents = loader.load()  
print(wiki_documents[0].page_content)
```
